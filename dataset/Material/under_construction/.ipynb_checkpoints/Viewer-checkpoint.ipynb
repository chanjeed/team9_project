ここから先はもとのjsonファイルを修正するプロセス

```python
#ファイル開く関数
def file_open(file_name):
    try:
        f = open(file_name,'r',encoding="utf-8")
    except:
        print("An exception occurred")
    json_data = json.load(f)
    return json_data
```

```python
#定数でファイルの位置を指定
#file_path =r"C:\Users\Sophia\Documents\Study\Deeplearning\DatasetStudy\cookpad_data-master\recipes.json"
file_path = r"../recipes.json"
```

```python
all_data = file_open(file_path)
```

```python
#中身のかたちはこんな感じ
"""
リスト形式で辞書型変数が10272個
{'ingredients': [{'amount': '5~6枚', 'name': '白菜'},
   {'amount': '1/2~1本', 'name': '白葱'},
   {'amount': '150g', 'name': '豚肉(バラ・薄切りなど)'},
   {'amount': '1本', 'name': '竹輪'},
   {'amount': '1枚', 'name': '油揚げ'},
   {'amount': '2玉', 'name': 'うどん(ゆで)'},
   {'amount': '2~3個', 'name': '餅'},
   {'amount': '500ml', 'name': '■水'},
   {'amount': '40ml', 'name': '■酒'},
   {'amount': '小さじ1', 'name': '■粉末だし'},
   {'amount': '大さじ1', 'name': '■醤油'},
   {'amount': '小さじ1/2', 'name': '■塩'}],
  'processes': ['うどん・餅以外の材料を食べやすい大きさに切る',
   '鍋か土鍋に水・酒・醤油・粉末だしを加えて火にかけて煮たてる',
   '白菜・豚肉を交互に重ねて最後に白葱をのせ、他の材料も入れる',
   '材料に火が通れば出来上がり♪'],
  'tags': ['野菜たっぷり 煮込みうどん'],
  'title': '❀我が家の白菜と豚肉の煮込みうどん❀'}
"""
all_data[0:3]
```

辞書型のファイルで処理するのがきついのでリスト化する？

```python
#空のデータフレーム作成
modified = pd.DataFrame(columns={"No","ingredients","tags","title","processes"})
modified = modified.loc[:,["No","title","tags","ingredients","processes"]]
modified
```

```python
#次の処理の変数を初期化する機能
def val_init():
    global ingredient_item 
    ingredient_item = ''
    global processes_item
    processes_item = ''
    global tags_item
    tags_item = ''
    return()
```

```python
#辞書のうちの必要箇所だけを切り取ってデータフレーム化
#めちゃ時間かかるのでこのセルは飛ばしてほしい


a = 0
#for list1 in range(len(all_data)):
for list1 in range(len(all_data)):
    val_init()
    a += 1
    #ingredientsを一つのStringへ
    for list2 in all_data[list1]["ingredients"]:
        ingredient_item = ingredient_item + list2["name"] + ","
    ingredient_item=ingredient_item.rstrip(",")
    #print (ingredient_item)
    
    #Processesを一つのStringへ
    for list3 in all_data[list1]["processes"]:
        processes_item = processes_item + list3+ ","
        processes_item=processes_item.rstrip(",")
    #print (processes_item)
    
    #tagsを一つのStringへ
    for list4 in all_data[list1]["tags"]:
        tags_item = tags_item + list4+ ","
        tags_item=tags_item.rstrip(",")
    #print (tags_item)
    modified = modified.append(pd.DataFrame({"No":list1, 
                                           "title":all_data[list1]["title"],
                                           "tags":tags_item,
                                           "ingredients": ingredient_item,
                                           "processes":processes_item}
                                          ,index = [str(list1)]))
    if a%100 == 0:
        print(a)
```

```python
#とりあえず一回csv出力
modified.to_csv(r"../modified.csv")
```
