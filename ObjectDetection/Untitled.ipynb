```python
import itertools
import math
```

```python
#入力画像サイズ
image_size = 300
# number of priors for feature map location (either 4 or 6)
aspect_ratios = [[2], [2, 3], [2, 3], [2, 3], [2], [2]]
num_priors = len(aspect_ratios)
variance =  [0.1, 0.2]
feature_maps = [38, 19, 10, 5, 3, 1]
#入力画像におけるAnchorのサイズ
min_sizes = [30, 60, 111, 162, 213, 264]
max_sizes = [60, 111, 162, 213, 264, 315]
#入力画像からの縮小率
steps = [8, 16, 32, 64, 100, 300]
clip = True
version = 'VOC'

mean = []
for k, f in enumerate(feature_maps):
    print('feature map:',f)
    for i,j in itertools.product(range(f), repeat=2):
        print('(i,j):',i,j)
        #feature mapサイズ
        f_k = image_size / steps[k]
        # unit center x,y
        cx = (j + 0.5) / f_k
        cy = (i + 0.5) / f_k    
        # aspect_ratio: 1
        # rel size: min_size
        s_k = min_sizes[k]/image_size
        mean += [cx, cy, s_k, s_k]
        print('(cx,cy,s_k):',cx,cy,s_k)
        
# aspect_ratio: 1
# rel size: sqrt(s_k * s_(k+1))
        s_k_prime = math.sqrt(s_k * (max_sizes[k]/image_size))
        mean += [cx, cy, s_k_prime, s_k_prime]
        print('(cx,cy,s_k):',cx,cy,s_k_prime)
        # rest of aspect ratios
        for ar in aspect_ratios[k]:
            print(ar)
            mean += [cx, cy, s_k*sqrt(ar), s_k/sqrt(ar)]
            mean += [cx, cy, s_k/sqrt(ar), s_k*sqrt(ar)]
# # back to torch land
# output = torch.Tensor(mean).view(-1, 4)
# if self.clip:
#     output.clamp_(max=1, min=0)
# output.required_grad = False
# return output

```
